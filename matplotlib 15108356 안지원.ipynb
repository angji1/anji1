## Matplotlib 사용

### Matplotlib 를 사용해서 cos 그래프 그려보기.

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.arange(380, step=20)
y = np.cos(x * 2 * np.pi / 360)

plt.plot(x, y)

plt.show()
```

#### x와 y의 값은 numpy에서 불러오고 짧은 코드식으로 cos 그래프를 그릴 수 있다는 것을 확인했다.

## Matplotlib의 응용

#### 위에서 간단한 cos그래프를 그리는 것만 해보았는데 여러가지 추가적인 함수를 사용해서 그래프를 추가적으로 그릴 수 있다.

### grid(격자) 나타내기

#### grid() 함수를 이용하면 격자를 그릴 수 있다.

```python
x = np.arange(380, step=20)
y = np.cos(x * 2 * np.pi / 360)

plt.plot(x, y)

plt.grid(True)

plt.show()
```

## 선 스타일 바꾸기

### 선 색상 바꾸기

#### 선의 색상은 plot() 함수에서 color 속성을 이용하여 변경한다.

```python
x = np.arange(380, step=20)
y = np.cos(x * 2 * np.pi / 360)

plt.plot(x, y, color = 'red')

plt.grid(True)

plt.show()
```

### 선의 두께는 linewidth 속성을 이용한다.

```python
x = np.arange(50, step=5)
y = x + 3
plt.plot(x, y, color = 'green' , linewidth = 5)

plt.grid(True)

plt.show()
```

### 선모양 바꾸는 함수는 linestyle 을 이용한다.

```python
x = np.arange(50, step=5)
y = x + 3

plt.plot(x, y, color = 'blue' , linewidth = 10, linestyle = "--")

plt.grid(True)

plt.show()
```

#### 이렇게 선의 스타일은 세가지를 함께 적용할 수 있다.

## Marker 스타일 바꾸기

#### marker는 선 위에 데이터 값을 표시하는 것을 말한다.

```python
x = np.arange(400, step=20)
y = np.sin(x * 2 * np.pi / 360)

plt.plot(x, y, color = 'red', linewidth = 0.5 , marker = 'v')

plt.grid(True)

plt.show()
```

```python
x = np.arange(380, step=20)
y = np.cos(x * 2 * np.pi / 360)

plt.plot(x, y, color = 'red', linewidth = 0.5, marker = 'o')

plt.grid(True)

plt.show()
```

#### 여러가지 모양으로 Mark 할 수 있다.

## 선 투명도 조절하기

#### plot() 함수에서 alpha 속성을 조절하면 된다.
#### 범위는 0 ~ 1 까지고 비교한 그래프를 보자.

```python
x = np.arange(380, step=20)
y = np.cos(x * 2 * np.pi / 360)

plt.plot(x, y, color = 'red', alpha = 1)

plt.grid(True)

plt.show()
```

```python
x = np.arange(380, step=20)
y = np.cos(x * 2 * np.pi / 360)

plt.plot(x, y, color = 'red', alpha = 0.6)

plt.grid(True)

plt.show()
```

```python
x = np.arange(380, step=20)
y = np.cos(x * 2 * np.pi / 360)

plt.plot(x, y, color = 'red', alpha = 0.2)

plt.grid(True)

plt.show()
```

#### 숫자가 작아질수록 점점 투명해지는 것을 확인할 수 있다.

## Label 표시하기

#### 그래프의 선이 무엇인지를 나타낼 수 있다.

```python
x = np.arange(380, step=20)
y = np.cos(x * 2 * np.pi / 360)

plt.plot(x, y, color = 'red')

plt.grid(True)

plt.legend(labels = str(1))

plt.show()
```

## 축의 label 붙이기

#### 각각 x축과 y축이 무엇을 나타내는지 표시할때 사용한다.

```python
x = np.arange(380, step=20)
y = np.cos(x * 2 * np.pi / 360)

plt.plot(x, y, color = 'red')

plt.grid(True)

plt.legend(labels = str(1))

plt.xlabel('S')

plt.ylabel('K')

plt.show()
```

## 축의 범위와 단위 정하기

### 우리가 그래프를 그릴때 X축과 Y축이 무엇을 나타내는지, 
### 그에따른 범위가 어디인지 어디까지 나타내야 하는지 생각해서 그래프를 그려야 한다. 
### 따라서 축의 범위를 지정하는 것은 중요하다.

```python
x = np.arange(380, step=20)
y = x/4
plt.plot(x, y, color = 'blue')

plt.grid(True)

plt.legend(labels = str('temp'))

plt.xticks([0,100,200,300], ['0 sec','100 sec','200 sec','300sec'])

plt.yticks([0,20,40,60,80], ['0K','20K','40K','60K','caution'])

plt.show()
```

#### 시간에 대한 온도를 나타낸 그래프이고 온도가 80K 가 넘으면 경고하도록 그래프를 그릴 수도 있다.

## 한번에 여러가지 그래프 그리기

### 작은 그래프 여러개를 한번에 그리려면 plot 대신 subplot 함수를 사용한다.
### 전체 행 수 전체 열 수 위치 인덱스를 입력해야한다.

```python
x = np.arange(380, step=20)
y = np.cos(x * 2 * np.pi / 360)

colors = ['red','green','blue','gray']

for i in range(1,5):
    
    plt.subplot(2,2,i)
    
    plt.plot(x,y, color = colors[i-1])
    
    plt.grid(True)

plt.show()
```

#### 이렇게 여러가지 그래프를 선의 스타일을 다르게 해서 그릴수도 있다.
